```python
df['sentiment'].replace('positive', 1, inplace=True)
df['sentiment'].replace('negative', 2, inplace=True)
```

```python
import nltk
nltk.download('punkt')
```

```python
# text processing
def data_processing(text):
    text = text.lower()
    text = re.sub('<br />', '', text) # break tag removal
    text = re.sub(r'https\S+|www\S+|http\S+', '', text, flags=re.MULTILINE)# removing weblinks
    text = re.sub(r'\@w+|\#', '', text) # removing 
    text = re.sub(r'[^\w\s]', '', text) # removing punctuations
    text_tokens = word_tokenize(text)
    filtered_text = [w for w in text_tokens if not w in stop_words] # removing stop words
    return " ".join(filtered_text)
```

```python
df.review = df['review'].apply(data_processing)
```

```python
duplicated_count = df.duplicated().sum()
print('Number of duplicated entries: ', duplicated_count)
```

```python
# dropping duplicated values
df = df.drop_duplicates('review')
```

```python
stemmer = PorterStemmer()
def stemming(data):
    text = [stemmer.stem(word) for word in data]
    return data
```

```python
df.review = df['review'].apply(lambda x: stemming(x))
```

```python
df['word count'] = df['review'].apply(no_of_words)
df.head()
```

```python
X = df['review']
Y = df['sentiment']
```

```python
vect = TfidfVectorizer()
X = vect.fit_transform(df['review'])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=.3, random_state=42)
```

```python
print('Size of X_train: ', X_train.shape)
print('Size of y_train: ', y_train.shape)
print('Size of X_test: ', X_test.shape)
print('Size of y_test: ', y_test.shape)
```

```python
# importing ML libraries
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import warnings
warnings.filterwarnings('ignore')
```

```python
# logistic regression model
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
logreg_pred = logreg.predict(X_test)
logreg_acc = accuracy_score(logreg_pred, y_test)
print('Test accuracy: {:.2f}%'.format(logreg_acc*100))
```

```python
print(confusion_matrix(y_test, logreg_pred))
print('\n')
print(classification_report(y_test, logreg_pred))
```

```python
# Multinomial NB model
mnb = MultinomialNB()
mnb.fit(X_train, y_train)
mnb_pred = mnb.predict(X_test)
mnb_acc = accuracy_score(mnb_pred, y_test)
print('Test accuracy: {:.2f}%'.format(mnb_acc*100))
```

```python
print(confusion_matrix(y_test, mnb_pred))
print('\n')
print(classification_report(y_test, mnb_pred))
```

```python
# SVC model
svc = LinearSVC()
svc.fit(X_train, y_train)
svc_pred = svc.predict(X_test)
svc_acc = accuracy_score(svc_pred, y_test)
print('Test accuracy: {:.2f}%'.format(svc_acc*100))
```

```python
print(confusion_matrix(y_test, svc_pred))
print('\n')
print(classification_report(y_test, svc_pred))
```

```python
# perform hyper parametric tuning for best parameters
from sklearn.model_selection import GridSearchCV
param_grid = {'C': [0.1, 1, 10, 100], 'loss':['hinge', 'sqaured_hinge']}
grid = GridSearchCV(svc, param_grid, refit=True, verbose=3)
grid.fit(X_train, y_train)
```

```python
print('Best cross validation score: {:.2f}'.format(grid.best_score_))
print('Best parameters: ', grid.best_params_)
```

```python
# SCV with best parameters
# SVC model
svc = LinearSVC(C=1, loss='hinge')
svc.fit(X_train, y_train)
svc_pred = svc.predict(X_test)
svc_acc = accuracy_score(svc_pred, y_test)
print('Test accuracy: {:.2f}%'.format(svc_acc*100))
```

```python
print(confusion_matrix(y_test, svc_pred))
print('\n')
print(classification_report(y_test, svc_pred))
```
